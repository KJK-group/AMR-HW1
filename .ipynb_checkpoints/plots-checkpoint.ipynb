```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np

plt.rcParams["figure.figsize"] = (8,6)
```

```python
ms = pd.read_csv(r'problem2_ms_4.csv')
of = pd.read_csv(r'problem2_of_4.csv')

plt.plot(ms.x, ms.y)
plt.plot(of.x, of.y)
plt.grid()
plt.legend(["/gazebo/model_states/pose[1]/position/x:y", "/odometry/filtered/pose/pose/position/x:y"])
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("problem1-position.svg", format="svg")
plt.show()
```

```python
def interpolate(inp, fi):
    i, f = int(fi // 1), fi % 1  # Split floating-point index into whole & fractional parts.
    j = i+1 if f > 0 else i      # Avoid index error.
    return (1-f) * inp[i] + f * inp[j]

len_input = len(of.x.values)
len_output = len(ms.x.values)
delta_len = (len_input - 1) / (len_output - 1)

of_interpolated = [
    [
        interpolate(of.timestamp.values, i * delta_len),
        interpolate(of.x.values, i * delta_len),
        interpolate(of.y.values, i * delta_len)
    ] for i in range(len_output)
]

"""
zlist = list(zip(ms.values.tolist(), of_interpolated))

(ms_time_n, ms_x_n, ms_y_n), (of_time_n, of_x_n, of_y_n) = zlist[0]
print(ms_time_n)
"""

error = np.array([
    [
        (of_time_n + ms_time_n) / 2,
        np.sqrt((of_x_n - ms_x_n)**2 + (of_y_n - ms_y_n)**2)
    ] for (ms_time_n, ms_x_n, ms_y_n), (of_time_n, of_x_n, of_y_n) in zip(ms.values.tolist(), of_interpolated)
])
```

```python
plt.plot(error[:,0], error[:,1])
plt.legend(["error"], loc='upper left')
plt.grid()
plt.legend(["position drift vs time"])
plt.xlabel("time")
plt.ylabel("drift")
plt.savefig("problem1-error.svg", format="svg")
```

```python
rho = pd.read_csv(r'problem1_rho_2.csv')

plt.plot(rho.timestamp, rho.error)
plt.grid()
plt.legend(["error (rho) vs time"])
plt.xlabel("time")
plt.ylabel("error (rho)")
plt.savefig("problem1-rho.svg", format="svg")
plt.show()
```

```python
yaw_data = pd.read_csv(r'yaw_angles.txt')
ms = pd.DataFrame(yaw_data, columns=['model_states_recceipt_time', 'model_states_yaw'])
of = pd.DataFrame(yaw_data, columns=['odometry_filtered_recceipt_time', 'odometry_filtered_yaw'])

nan_value = float("NaN")
of.replace(" ", nan_value, inplace=True)
of.dropna(inplace=True)

of = of.astype(float)
of.dtypes
```

```python
plt.plot(ms.model_states_recceipt_time, ms.model_states_yaw)
plt.plot(of.odometry_filtered_recceipt_time, of.odometry_filtered_yaw)
plt.grid()
plt.xlabel("time")
plt.ylabel("yaw (radians)")
plt.legend(["/gazebo/model_states", "/odometry/filtered"], loc='upper left')
plt.savefig("problem2-yaw.svg", format="svg")
plt.show()
```

```python
lower_bound = 33.5
upper_bound = 36

ms_slice = np.array([
    [
        time_n,
        yaw_n
    ] for time_n, yaw_n in ms.values if time_n > lower_bound and time_n < upper_bound
])

of_slice = np.array([
    [
        time_n,
        yaw_n
    ] for time_n, yaw_n in of.values if time_n > lower_bound and time_n < upper_bound
])
#print(ms_slice[:,0])

plt.plot(ms_slice[:,0], ms_slice[:,1])
plt.plot(of_slice[:,0], of_slice[:,1])
plt.grid()
plt.xlabel("time")
plt.ylabel("yaw (radians)")
plt.legend(["/gazebo/model_states", "/odometry/filtered"], loc='upper left')
plt.savefig("problem2-yaw-zoomed.svg", format="svg")
plt.show()
```
